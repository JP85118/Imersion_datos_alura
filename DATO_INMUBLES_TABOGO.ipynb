PROMEDIO DE AREA DE LOS BARRIOS

```python
#Los 10 barrios con el mayor promedio de area
promedio_barrios = pd.DataFrame(inmuebles.groupby('Barrio').Area.mean()).round(2)
promedio_barrios.sort_values('Area',ascending=False,inplace=True)
promedio_barrios.head(10)
```

```python
#Grafico del top 10 de barrios de mayor a menos area de inmuebles en bogota
promedio_barrios = inmuebles.groupby('Barrio')['Area'].mean().sort_values(ascending=False).head(10)
promedio_barrios.plot.bar()
```

```python
conteo = inmuebles.Area.count()
print(f'El conteo es de: {conteo}')
mediana = inmuebles.Area.median()
print(f'La mediana del area es: {mediana}')
val_min = inmuebles.Area.min()
print(f'El valor minino de area es:{val_min}')
val_max = inmuebles.Area.max()
print(f'El valor maximo de area es:{val_max}')
```

```python
inmuebles.Valor[0].split()
```

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles[['Precio', 'Barrio']]
```

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```
